```python
import numpy as np

e3 = np.array([1/3, 2/3, 2/3])                       
e1 = np.array([2/np.sqrt(5), -1/np.sqrt(5), 0])      
e2 = np.cross(e3, e1)                                

Q = np.column_stack((e1, e2, e3))                    


A1 = np.array([[0, -1, 0],
               [1,  0, 0],
               [0,  0, 1]])

A = Q @ A1 @ Q.T

np.set_printoptions(suppress=True, precision=10)
print("e2 = \n",e2)

print("Q =\n", Q)
print("\nA =\n", A)

print("\nA * 9 ≈\n", 9*A)

axis = np.array([1,2,2])
print("\nA @ axis =\n", A @ axis)
```

```python
e3 = np.array([1, 1, 1])                       
e1 = np.array([-1, 1, 0])      
e2 = np.cross(e3, e1)                                

# Normalise
e1 = e1 / np.linalg.norm(e1)
e3 = e3 / np.linalg.norm(e3)
e2 = e2 / np.linalg.norm(e2)
                         
Q = np.column_stack((e1, e2, e3))
print(np.allclose(Q.T, np.linalg.inv(Q))) 
```
